```python
import pandas as pd

data = pd.read_csv("Indian automoble buying behavour study 1.0.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
# here we can see all the columns contain nonnull values. so we dont have to worry about null values.

#1. Lets see if people with more salary buy more expensive EV or not?

import matplotlib.pyplot as plt
x = data['Salary'].values
y = data['Price'].values

plt.scatter(x, y, color='blue', alpha=0.5)
plt.xlabel("Salary of an individual")
plt.ylabel("Price of an EV")
plt.show()
```

```python
# As we can see there is no clear relationship between salary and price. Lets check with the total salary and price
x = data['Total Salary'].values
y = data['Price'].values

plt.scatter(x, y, color='blue', alpha=0.5)
plt.xlabel("Total salary of a household")
plt.ylabel("Price of an EV")
plt.show()
```

```python
# here also we cannot find any insight

# 2. Lets see distribution of the salaries and total salaries
x = data['Salary'].values
plt.hist(x, bins='auto', color='blue', alpha=0.7)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Distribution of Salaries')
plt.show()
```

```python
# as we can see the salaries are normally distributed
# lets also check the distirbution of other columns as well
x = data['Total Salary'].values
plt.hist(x, bins='auto', color='blue', alpha=0.7)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Distribution of Total Salary')
plt.show()
```

```python
x = data['Price'].values
plt.hist(x, bins='auto', color='blue', alpha=0.7)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Distribution of Price')
plt.show()
```

```python
# from the above graph we can see most of the cars are priced around fixed range

# 3. Lets check which age's individual buys which priced cars so we can better target customer age wise.

x = data['Age'].values
y = data['Price'].values
plt.scatter(x,y)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Distribution of Total Salary')
plt.show()
```

```python
# 4. Lets check people with or without house loan buys how expensive car

grouped_df = data.groupby('Personal loan').agg({'Salary': 'mean', 'Price': 'mean'})
# Plot the average salary and average car price for each marital status
plt.figure(figsize=(10, 6))
grouped_df.plot(kind='bar')
plt.title('Average Salary and Car Price by Marital Status')
plt.xlabel('Personal loan')
plt.ylabel('Amount')
plt.legend(['Average Salary', 'Average Car Price'])
plt.xticks(rotation=0)
plt.show()
```

```python
# here we can see the people with Perosnal loan or without loan tend to spend same proportion of money in buying EV

# lets check with respect to house loan also.
grouped_df = data.groupby('House Loan').agg({'Salary': 'mean', 'Price': 'mean'})
# Plot the average salary and average car price for each marital status
plt.figure(figsize=(10, 6))
grouped_df.plot(kind='bar')
plt.title('Average Salary and Car Price by Marital Status')
plt.xlabel('Personal loan')
plt.ylabel('Amount')
plt.legend(['Average Salary', 'Average Car Price'])
plt.xticks(rotation=0)
plt.show()
```

```python
# here we can see people with house loan tend to spend less proportion of their salary on EV than people without the House Loan

# 5. Lets check whether number of dependents impact buying the high priced EV
# lets check with respect to house loan also.
grouped_df = data.groupby('No of Dependents').agg({'Salary': 'mean', 'Price': 'mean'})
# Plot the average salary and average car price for each marital status
plt.figure(figsize=(10, 6))
grouped_df.plot(kind='bar')
plt.title('Average Salary and Car Price by Marital Status')
plt.xlabel('Personal loan')
plt.ylabel('Amount')
plt.legend(['Average Salary', 'Average Car Price'])
plt.xticks(rotation=0)
plt.show()
```

```python
# as we can see from the above graphic people with no dependents tend to spend more money of their salary on EV

# 6. Lets check the same with respect to their marital status, education and other factors

for i in ['Education','Marrital Status','Profession']:
    grouped_df = data.groupby(i).agg({'Salary': 'mean', 'Price': 'mean'})
    # Plot the average salary and average car price for each marital status
    plt.figure(figsize=(10, 6))
    grouped_df.plot(kind='bar')
    plt.title('Average Salary and Car Price by Marital Status')
    plt.xlabel('Personal loan')
    plt.ylabel('Amount')
    plt.legend(['Average Salary', 'Average Car Price'])
    plt.xticks(rotation=0)
    plt.show()
```

```python
# 7. Lets check the distribution of the car models
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Make')
plt.title('Vehicle Make Analysis')
plt.xlabel('Make')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
